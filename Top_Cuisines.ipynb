```python
#import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data set
url="https://drive.google.com/file/d/1dKZxapT3xLzOTOpy1LCyqctL8YEvzp4Y/view?usp=drive_link"
file_id = url.split("/")[5]
url= f"https://drive.google.com/uc?id={file_id}"
df=pd.read_csv(url)
```

```python
print(df.shape)
print(df.columns)
```

```python
print(df['Cuisines'])
```

```python
cuisine_count=df['Cuisines'].str.split(', ').explode('Cuisines').value_counts()
print(cuisine_count)
top=cuisine_count.head(3)
print("top three cuisines are:",top)
```

```python
colors=['green','orange','red']
plt.bar(top.index,top.values,color=colors)
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.title("Top 3 Cuisines")
plt.show()
```

```python
total_restaurant=len(df)
print(total_restaurant)
top_cuisine=cuisine_count.head()
percentage=(top_cuisine/total_restaurant)*100
print("The market share of top 5 cuisines are : ",percentage)
```

```python
plt.bar(top_cuisine.index,percentage.values,color=colors)
plt.xlabel('Cuisine Name')
plt.ylabel('Percentage')
plt.title("Top 5 Cuisines %")
plt.figure(figsize=(12,6))
plt.show()
```
